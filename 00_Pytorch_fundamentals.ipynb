### Random tensors

Why random tensors ?

Random tensors are important because the way many neural networks learn is that they start with tensors full of random numbers and then adjust those random numbers to represent the data better.

`start with random numbers -> look at data -> update random numbers -> look at data -> update random numbers`

torch random tensors - "https://docs.pytorch.org/docs/stable/generated/torch.rand.html"

```python
# Create a random tensor of size (3,4)
random_tensor = torch.rand(3,4)
random_tensor
```

```python
# Create a random tensor with similar shape to an image tensor
# "size" can be eliminated without any problem
random_image_tensor = torch.rand(size=(3, 224,224)) #  color channels (R,G,B), height, width
random_image_tensor
```

### Tensors datatypes

**Note:** Tensor datatypes is one of the 3 big errors we'll run into with pyTorch & deep learning:
1. Tensors not right datatype
2. Tensors not right shape
3. Tensors not on the right device  

```python
tens_1 = torch.tensor([3.0, 4.0, 5.0],
                      dtype=None,   # what is the datatype of the tensor(e.g. float32, float16)
                      device=None,  # what device is our tensor on (e.g. cpu or cuda)
                      requires_grad=False) # whether or not to track gradients with this tensor operations

tens_2 = torch.tensor([1.0,2.0,3.0], dtype=torch.float16)
```

```python
tens_1.dtype
```

```python
tens_1*tens_2
```

### Manipulating Tensors (tensor operations)

1. Addition
2. Subtraction
3. Element-wise multiplication
4. Division
5. Matrix multiplication

```python
tensor_a = torch.rand(2,3)
tensor_b = torch.rand(2,3)
```

```python
torch.mul(tensor_a, tensor_b) # Element wise multiplication (also can be like: tensor_a * tensor_b)
```

```python
%%time
tensor_a @ tensor_b.T # Matrix multiplication (It is the same as: matmul(tensor_a,tensor_b.T))
```

### Squeezing and unsqueezing
1. Squeeze --> remove all 1 dimensions from a tensor
2. Unsqueeze --> add all 1 dimension to a target tensor
3. Permute --> Return a view of the input with dimensions permuted (swapped) in a certain way

```python
y = torch.arange(10,201,10)
```

```python
y
```

```python
r = y.view(1,20)
```

```python
r.size()
```

```python
r = r.squeeze()
```

```python
r.size()
```

```python
rs = r.unsqueeze(dim=1)
```

```python
rs.shape
```

### Putting tensors(and models) on the GPU

Using GPU --> faster computations



```python
tensor = torch.tensor([1,2,3], device="cpu") # by default it is on CPU
tensor, tensor.device
```

```python
# Move the tensor to GPU (if available)
device = "cuda" if torch.cuda.is_available() else "cpu"

tensor_on_gpu = tensor.to(device)
tensor_on_gpu
```

### Moving tensors back to the CPU

- As Numpy doesn't work work GPU, we can't transform a tensor on GPU to Numpy

- To do so, we need first to set the tensor to CPU


```python
tensor_on_cpu = tensor_on_gpu.cpu().numpy()
tensor_on_cpu
```
